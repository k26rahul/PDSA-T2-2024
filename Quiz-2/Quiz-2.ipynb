**Statement 2: We can construct back $ T $ if only Preorder traversal is known.**

- Preorder lists nodes as root, left subtree, right subtree.
- For a general tree, preorder is insufficient.
- For BSTs, preorder allows unique construction by dividing nodes into subtrees.

**Statement 3: We can construct back $ T $ if only Postorder traversal is known.**

- Postorder lists nodes as left subtree, right subtree, root.
- For BSTs, postorder allows unique construction by identifying the root and dividing nodes.

**Statement 4: We can construct back $ T $ if any one pair of the three Preorder, Inorder, Postorder traversals is known.**

- Any two traversals (preorder and inorder, or postorder and inorder) allow reconstruction:

  - **Postorder and Inorder**: Root from postorder, structure from inorder.
  - **Preorder and Postorder**: For BSTs, this pair uniquely determines the structure.
