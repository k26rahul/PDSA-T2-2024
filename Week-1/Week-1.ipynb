# Computing gcd

```python
def gcd(m, n):
  common_factors = []
  for i in range(1, min(m, n)+1):
    if m % i == 0 and n % i == 0:
      common_factors.append(i)

  return common_factors[-1]


print(gcd(8, 12))
print(gcd(18, 25))
```

### Computing gcd - Eliminate the list: `common_factors = []`

```python
def gcd(m, n):
  for i in range(1, min(m, n)+1):
    if m % i == 0 and n % i == 0:
      largest_common_factor = i

  return largest_common_factor


print(gcd(8, 12))
print(gcd(18, 25))
```

### Computing gcd - Better Way

$ m = xd, n = yd $

$ m - n = (x - y)d $

$ d $ also divides $ m - n $

```python
def gcd(m, n):
  a, b = max(m, n), min(m, n)

  print(f'{a=}, {b=}')

  if a % b == 0:
    return b
  else:
    return gcd(b, a-b)


# print(gcd(8, 12))
# print(gcd(18, 25))
# print(gcd(45, 75))
print(gcd(90, 1700))
```
